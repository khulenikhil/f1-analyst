```python
schedules = None
for i in range(2024,2018,-1):
    with fastf1.Cache.disabled():
        if schedules is None:
            schedules = pd.DataFrame(fastf1.get_event_schedule(i))
        else:
            schedules = pd.concat([schedules, pd.DataFrame(fastf1.get_event_schedule(i))])
        print(f"Event schedule fetched for {i}")
else:
    print(schedules.shape)
```

```python
%%sql
CREATE TABLE IF NOT EXISTS
    sessions
AS
    WITH session_1 AS (
        SELECT
            "RoundNumber" as round_number,
            "Country" as country,
            "Location" as location,
            CASE WHEN "OfficialEventName" = '' THEN "EventName" ELSE "OfficialEventName" END as official_event_name,
            "EventDate" as event_date,
            "EventName" as event_name,
            "EventFormat" as event_format,
            "Session1" as session_type,
            "Session1DateUtc" as session_date
        FROM bronze_events
    ), session_2 AS (
        SELECT
            "RoundNumber" as round_number,
            "Country" as country,
            "Location" as location,
            CASE WHEN "OfficialEventName" = '' THEN "EventName" ELSE "OfficialEventName" END as official_event_name,
            "EventDate" as event_date,
            "EventName" as event_name,
            "EventFormat" as event_format,
            "Session2" as session_type,
            "Session2DateUtc" as session_date
        FROM bronze_events
    ), session_3 AS (
        SELECT
            "RoundNumber" as round_number,
            "Country" as country,
            "Location" as location,
            CASE WHEN "OfficialEventName" = '' THEN "EventName" ELSE "OfficialEventName" END as official_event_name,
            "EventDate" as event_date,
            "EventName" as event_name,
            "EventFormat" as event_format,
            "Session3" as session_type,
            "Session3DateUtc" as session_date
        FROM bronze_events
    ), session_4 AS (
        SELECT
            "RoundNumber" as round_number,
            "Country" as country,
            "Location" as location,
            CASE WHEN "OfficialEventName" = '' THEN "EventName" ELSE "OfficialEventName" END as official_event_name,
            "EventDate" as event_date,
            "EventName" as event_name,
            "EventFormat" as event_format,
            "Session4" as session_type,
            "Session4DateUtc" as session_date
        FROM bronze_events
    ), session_5 AS (
        SELECT
            "RoundNumber" as round_number,
            "Country" as country,
            "Location" as location,
            CASE WHEN "OfficialEventName" = '' THEN "EventName" ELSE "OfficialEventName" END as official_event_name,
            "EventDate" as event_date,
            "EventName" as event_name,
            "EventFormat" as event_format,
            "Session5" as session_type,
            "Session5DateUtc" as session_date
        FROM bronze_events
    )
    
    SELECT
        *
    FROM
        (
            SELECT * FROM session_1
            UNION ALL
            SELECT * FROM session_2
            UNION ALL
            SELECT * FROM session_3
            UNION ALL
            SELECT * FROM session_4
            UNION ALL
            SELECT * FROM session_5
        )
    ORDER BY
        session_date ASC,
        event_date ASC;
```

```python
%%sql

SELECT * FROM sessions ORDER BY event_date ASC, session_date ASC;
```

```python
%%sql

SELECT
    round_number,
    EXTRACT(year FROM event_date) as "year",
    location,
    official_event_name,
    event_name,
    event_date,
    event_format
FROM
    sessions
GROUP BY
    round_number,
    location,
    official_event_name,
    event_name,
    event_date,
    event_format
ORDER BY
    "year", round_number;
```
